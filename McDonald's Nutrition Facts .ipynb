```python
# slicing to get sandwich only
sandwich = df[df['Category']=="Chicken & Fish"][:4]
sandwich
```

```python
# get the crispy chicken sandwich only
sandwich_crispy = sandwich[sandwich['Item'].str.contains('Crispy')]
sandwich_crispy
```

```python
# get the grilled chicken sandwich only
sandwich_grilled = sandwich[sandwich['Item'].str.contains('Grilled')]
sandwich_grilled
```

```python
# As chicken nutrition facts only have Calories, Protein, and Fat. We are going to check those three.
# sandwich calories comparison
sandwich_cal = pd.DataFrame({'Grilled': [sandwich_grilled["Calories"].mean()], 
                                'Crispy': [sandwich_crispy["Calories"].mean()]
                               })
sandwich_cal.plot.bar()
plt.ylabel('Calories')
plt.xlabel('Sandwich Type')
plt.show()

print("Grilled has more calories than crispy: " + 
      str(sandwich_grilled["Calories"].mean() > sandwich_crispy["Calories"].mean()))
```

```python
# sandwich protein comparison
sandwich_pro = pd.DataFrame({'Grilled': [sandwich_grilled["Protein"].mean()], 
                                'Crispy': [sandwich_crispy["Protein"].mean()]
                               })
sandwich_pro.plot.bar()
plt.ylabel('Protein')
plt.xlabel('Sandwich Type')
plt.show()

print("Grilled has more protein than crispy: " + 
      str(sandwich_grilled["Protein"].mean() > sandwich_crispy["Protein"].mean()))
```

```python
# sandwich protein comparison
sandwich_pro = pd.DataFrame({'Grilled': [sandwich_grilled["Total Fat"].mean()], 
                                'Crispy': [sandwich_crispy["Total Fat"].mean()]
                               })
sandwich_pro.plot.bar()
plt.ylabel('Total Fat')
plt.xlabel('Sandwich Type')
plt.show()

print("Grilled has more Total Fat than crispy: " + 
      str(sandwich_grilled["Total Fat"].mean() > sandwich_crispy["Total Fat"].mean()))
```

```python
# As egg nutrition facts only have Calories, Protein, Carbs, and Fat. We are going to check those four.
# get egg
egg = df[df['Item'].str.contains('Egg')]
egg
```

```python
# filter only white egg
white_egg =  df[df['Item'].str.contains('White')]
white_egg
```

```python
# exclude white egg
whole_egg = egg[~df['Item'].str.contains('White')]
whole_egg
```

```python
#Nutrition check
#calories comparison
egg_cal = pd.DataFrame({'White': [white_egg["Calories"].mean()], 
                            'Whole': [whole_egg["Calories"].mean()]
                        })

egg_cal.plot.bar()
plt.ylabel('Total Fat')
plt.xlabel('Egg Type')
plt.show()

print("White egg has more calories than whole egg: " + 
      str(white_egg["Calories"].mean() > whole_egg["Calories"].mean()))
```

```python
#protein comparison
egg_pro = pd.DataFrame({'White': [white_egg["Protein"].mean()], 
                            'Whole': [whole_egg["Protein"].mean()]
                        })

egg_pro.plot.bar()
plt.ylabel('Protein')
plt.xlabel('Egg Type')
plt.show()

print("White egg has more protein than whole egg: " + 
      str(white_egg["Protein"].mean() > whole_egg["Protein"].mean()))
```

```python
nRowsRead = 1000 # specify 'None' if want to read whole file
data = pd.read_csv('menu.csv', delimiter=',', nrows = nRowsRead)
data.dataframeName = 'menu.csv'
nRow, nCol = data.shape
print(f'There are {nRow} rows and {nCol} columns')
```

```python
data.head(5)
```

```python
data.shape
```

```python
plt.figure(figsize=(10, 4), dpi=100)
menu_category = data.Category.value_counts()
menu_category.plot.bar(color = ['red','pink','pink','pink','pink','pink','pink','pink','pink'])
plt.title("Number of Menu Items for each Food Category")
plt.ylabel("Count")
plt.xlabel("Menu Category")
plt.xticks(rotation=45)
plt.show()
```

```python
data.groupby('Category')['Trans Fat'].mean() 
```

### Fats and Carbohydrates

```python
plt.figure(figsize=(10, 4), dpi=100)

correlation = data['Calories'].corr(data['Calories from Fat'])
plt.scatter(data.Calories, data['Calories from Fat'], color='black')
plt.text(500,450,'r = {}'.format(round(correlation,2)))
plt.xlabel("Calories")
plt.ylabel("Calories from Fat")
plt.title("Relationship between Calories and Calories from Fat")
plt.show()
```

```python
data['saturated_cholesterol'] = data['Saturated Fat']/data['Cholesterol']*100
```

```python
saturated_cholesterol = data.groupby('Category')['saturated_cholesterol'].mean().dropna().nlargest(5)
```

```python
coffee_tea = data[data.Category == 'Coffee & Tea']
coffee_tea.groupby('Item')['saturated_cholesterol'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 4), dpi=100)
saturated_cholesterol.sort_values(ascending=False).plot.bar(color = 'orange')
plt.title("Top 5 food categories with the highest % Saturated Fat")
plt.ylabel("Percentage")
plt.xlabel("Menu Category")
plt.xticks(rotation=45)
plt.show();
```

### Vitamins & Minerals

```python
beverage = data[data.Category == 'Beverages']
beverage.groupby('Item')['Vitamin C (% Daily Value)'].mean().sort_values()
```

```python
shakes = data[data.Category == 'Smoothies & Shakes']
shakes.groupby('Item')['saturated_cholesterol'].mean().sort_values()
```

```python
data.groupby('Category')['Vitamin A (% Daily Value)'].mean()
```

```python
data.groupby('Category')['Vitamin C (% Daily Value)'].mean()
```

```python
iron = data.groupby('Category')['Iron (% Daily Value)'].mean()
calcium = data.groupby('Category')['Calcium (% Daily Value)'].mean()
```

```python
plt.figure(figsize=(10, 4), dpi=100)

category = iron.index
Iron = iron.values
Calcium = calcium.values

X_axis = np.arange(len(category))

plt.bar(X_axis - 0.2, Iron, 0.4, label = 'Iron', color = 'pink')
plt.bar(X_axis + 0.2, Calcium, 0.4, label = 'Calcium', color = 'black')

plt.xticks(X_axis, category)
plt.xlabel("Menu Categories")
plt.title("Distribution of Iron and Calcium in McDonald's Foods")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

Mcdonald's beverages are very low in Calcium and Iron.

```python
plt.figure(figsize=(10, 4), dpi=100)
plt.scatter(data['Total Fat (% Daily Value)'], data['Carbohydrates (% Daily Value)'], color='k')
correlation = data['Total Fat (% Daily Value)'].corr(data['Carbohydrates (% Daily Value)'])
plt.text(125,20,'r = {}'.format(round(correlation,2)))
plt.xlabel("% Total Fat")
plt.ylabel("% Total Carbohydrate")
plt.title("Relationship between % Total Fat and % Total Carbohydrate")
plt.show()
```

There is a strong relationship between Calories and Calories with fat. It means food items with high caloric content tend to also have a high calories with fat content.

+ Generally, McDonalds foods have low trans fat which is good.

### Conclusion

+ The McDonald's Menu list and the nutritional value of each food category in the list depends on  nutritional needs.
+  A lot of Mcdonald's foods are rich in Vitamins (Vitamin A & C) especially Salads. Also, most foods are low in carbohydrate with little or no trans-fat which is a plus. However, a lot of foods are high in Saturated fat, hence, it is important to check for foods with low Saturated fat (>10%) in case you are cutting on your fat intake.The analysis also covers essential minerals such as Calcium and Iron and it was noted that foods in the menu containing milk or animal protein are very rich in calcium.
